```python
class FileSystemControl():
    meta_data = "file_system_ingestion_meta_data.xlsx"
    def __init__(self, sheet_name = ''):
        self.sheet_name = sheet_name

    def preprocess_columns(self, df = pd.DataFrame()):
        try:
            for col in df.columns:
                if df[col].dtype == object:
                    df[col] = df[col].str.strip()
            return df
        except:
            print(traceback.format_exc())
            return 'preprocess_column failed'
    def fetch_active_file_system(self):
        fetch_fs = pd.read_excel(FileSystemControl.meta_data, sheet_name = self.sheet_name)
        fetch_fs = fetch_fs[fetch_fs['isActive']==1]
        fetch_fs = self.preprocess_columns(fetch_fs)

        return fetch_fs
    def ingestion_process(self):
        file_systems_df = self.fetch_active_file_system()
        for row in file_systems_df.itertuples():
            if getattr(row,'file_systems') == 'Azure ADLS':
                # print("We are in azure")
                account_details = 'azure_accounts_details'
                azure_files = 'azure_files_details'
                az_fs = AzureFileSystem(account_sheet = account_details, files_sheet = azure_files)
                az_fs.azure_file_ingestion()
            elif getattr(row,'file_systems') == 'AWS S3':
                print("We are in AWS")
            elif getattr(row,'file_systems') == 'GCP Storage':
                print("We are in GCP")
            else:
                print("Random File System")
```

```python
if __name__ == '__main__':
    file_systems = 'filesystem_details'
    file_ingestion = FileSystemControl(sheet_name = file_systems)
    file_ingestion.ingestion_process()
```
